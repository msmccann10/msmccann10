```python
mortgage_df = pd.read_csv("/content/drive/MyDrive/Data/mortgages.csv")
mortgage_df.head()
```

```python
mortgage_df['Interest Rate'].value_counts()
```

```python
interest_filter = mortgage_df['Interest Rate'] == 0.03
interest_filter.head()
```

```python
mortgage_df[interest_filter].head()

#same as writing out mortgage_df[mortgage_df['Interest Rate'] == 0.03].head()
```

```python
mortgage_df.loc[interest_filter, :].head()

#same as mortgage_df.loc[mortgage_df['Interest Rate'] == 0.03, :].head()
```

Combine Filters using 
& and
| or
^ exclusive or
~ not

```python
mortgage_filter = mortgage_df['Mortgage Name'] == '30 Year'

mortgage_df.loc[mortgage_filter & interest_filter, :]
```
